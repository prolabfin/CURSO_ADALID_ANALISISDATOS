Para resolver las diferentes partes del ejercicio, podemos utilizar bibliotecas de Python como NumPy, Matplotlib y Seaborn para realizar los cálculos y crear gráficos.

a) Tabla de Frecuencia y Diagrama de Barras:

```python
import numpy as np
import matplotlib.pyplot as plt

metros = np.array([1, 2, 3, 4, 5, 6, 7, 8])
niños = np.array([2, 6, 10, 5, 10, 3, 2, 2])

frecuencia_absoluta = niños
frecuencia_relativa = frecuencia_absoluta / np.sum(frecuencia_absoluta)
frecuencia_acumulada = np.cumsum(frecuencia_relativa)

tabla_frecuencia = np.column_stack((metros, niños, frecuencia_absoluta, frecuencia_relativa, frecuencia_acumulada))
print("Tabla de Frecuencia:")
print(tabla_frecuencia)

# Diagrama de Barras para Frecuencia Absoluta
plt.bar(metros, frecuencia_absoluta)
plt.xlabel("Metros")
plt.ylabel("Frecuencia Absoluta")
plt.title("Diagrama de Barras - Frecuencia Absoluta")
plt.show()

# Diagrama de Barras para Frecuencia Relativa
plt.bar(metros, frecuencia_relativa)
plt.xlabel("Metros")
plt.ylabel("Frecuencia Relativa")
plt.title("Diagrama de Barras - Frecuencia Relativa")
plt.show()

# Diagrama de Barras para Frecuencia Acumulada
plt.bar(metros, frecuencia_acumulada)
plt.xlabel("Metros")
plt.ylabel("Frecuencia Acumulada")
plt.title("Diagrama de Barras - Frecuencia Acumulada")
plt.show()
```

```python
# b) Mediana, media aritmética, moda y cuartiles.
from scipy import stats

mediana = np.median(metros)
media_aritmetica = np.mean(metros)
moda = stats.mode(metros)
cuartiles = np.percentile(metros, [25, 50, 75])

print("Mediana:", mediana)
print("Media Aritmética:", media_aritmetica)
print("Moda:", moda.mode)
print("Cuartiles:", cuartiles)
```

```python
# c) Varianza y desviación Estándar.
varianza = np.var(metros)
desviacion_estandar = np.std(metros)

print("Varianza:", varianza)
print("Desviación Estándar:", desviacion_estandar)
```

```python
# D) ¿Entre qué dos valores se encuentra, como mínimo, el 75% de las
#observaciones?

percentil_75 = np.percentile(metros, 75)
valor_minimo_75 = np.min(metros[metros >= percentil_75])

print("Valor mínimo para el 75% de las observaciones:", valor_minimo_75)
```
