```python
# Leer csv correspondiente. ● Renombrar campos. 

import pandas as pd # Importacion estandar de la libreria Pandas
import numpy  as np # Importacion estandar de la libreria NumPy


df = pd.read_csv("/content/M4-AP03_casasboston.csv")
df.info()
df.head(5)
```

```python
#  Un histograma es útil para ver la distribución de una variable, es decir, nos permite ver los valores más comunes. 
#Para el caso que se presenta se desea ver la distribución de la cantidad media de habitaciones en el estudio realizado.

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Renombrar la columna 'RM' a 'habitaciones'
df = df.rename(columns={'RM': 'habitaciones'})
df.head(5)

```

```python
# Cálculo de la cantidad media de habitaciones
media_habitaciones = np.mean(df['habitaciones'])

# Creación del histograma
plt.hist(df['habitaciones'], bins=10)  # Ajusta el número de bins según tus necesidades
plt.xlabel('Cantidad de habitaciones')
plt.ylabel('Frecuencia')
plt.title('Distribución de la cantidad de habitaciones')
plt.axvline(media_habitaciones, color='r', linestyle='dashed', linewidth=1, label='Media')  # Línea vertical para la media
plt.legend()
plt.show()
```

```python
#Dispersión: sirve para representar la relación entre dos variables. Para nuestro caso se desea ver la relación entre índice de crimen y el valor mediano de las casas. 
#En otras palabras, ¿el índice de crimen afecta el valor medio de las casas?

import numpy as np
import matplotlib.pyplot as plt

# Extracción de los valores de CRIM y MEDV
x = df['CRIM'].values
y = df['MEDV'].values

# Creación del gráfico de dispersión
plt.scatter(x, y)
plt.xlabel('Índice de crimen')
plt.ylabel('Valor medio de las casas')
plt.title('Relación entre el índice de crimen y el valor medio de las casas')
plt.show()


```

Segun lo estudiado no existe una relacion clara entre estas dos variables solo que en el rango de indice de crimen  [0-20] el valor medio de las casas se mantiene muy alto mientras este indice es 0 a medida que aumenta el indice de crimen el valor medio de las casas es menos y mas disperso.  No existe una correlacion ni valor directo .

Barras: Un gráfico de barras es útil para comparar una variable
entre distintos grupos o categorías. Por ejemplo, si quisiéramos
observar el valor medio de cada ciudad. Para nuestro caso se
desea agrupar por ciudad y toma la media del valor mediano.
Gráfica las 10 primeras ciudades.

```python
import numpy as np
import matplotlib.pyplot as plt

# Proporción de acres de negocios no minoristas por ciudad/pueblo

# Agrupación por ciudad y cálculo del valor medio de INDUS
media_por_ciudad = df.groupby('TOWN')['INDUS'].mean()

# Obtención de las 10 primeras ciudades y sus valores medios
top_10_ciudades = media_por_ciudad.head(10)

# Creación del gráfico de barras
plt.bar(top_10_ciudades.index, top_10_ciudades)
plt.xlabel('Ciudad')
plt.ylabel('Valor medio de INDUS')
plt.title('Valor medio de INDUS en las 10 primeras ciudades')
plt.xticks(rotation=45)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Agrupación por ciudad y cálculo del valor medio de •	PTRATIO: Proporción de alumnos por profesor por ciudad/pueblo
media_por_ciudad2 = df.groupby('TOWN')['PTRATIO'].mean()

# Obtención de las 10 primeras ciudades y sus valores medios
top_10_ciudades2 = media_por_ciudad2.head(10)

# Creación del gráfico de barras
plt.bar(top_10_ciudades2.index, top_10_ciudades2)
plt.xlabel('Ciudad')
plt.ylabel('Valor medio de alumnos por profesor por ciudad pueblo')
plt.title('Valor medio de alumnos por profesor por ciudad pueblo en las 10 primeras ciudades')
plt.xticks(rotation=45)
plt.show()
```
