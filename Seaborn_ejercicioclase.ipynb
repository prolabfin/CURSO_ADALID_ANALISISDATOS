```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
import seaborn as sns


```

```python
sns.get_dataset_names()
```

```python
taxis = sns.load_dataset('titanic')

taxis
```

```python
x = taxis["distance"]
y = taxis["fare"]

sns.jointplot(x=x, y=y, data=taxis, kind="scatter", height=6, ratio=5, space=0.2, xlim=None, ylim=None, joint_kws=None, marginal_kws=None)

```

```python
# E múltiples tipos de gráficos para proporcionar rápidamente resúmenes informativos de un conjunto de datos. Uno, jointplot(), se enfoca en una sola relación. 
#Traza la distribución conjunta entre dos variables junto con la distribución marginal de cada variable:

#sns.jointplot(data=taxis, x="distance", y="fare", hue="tip",xlim=[0:60], ylim=[0:200])
sns.jointplot(data=taxis, x="distance", y="fare", hue="tip")

```

```python
# Se importa la librería Seaborn para hacer gráficos estadísticos
import seaborn as sns

# Se utiliza la función pairplot de Seaborn para crear un gráfico de dispersión de todas las variables numéricas en un conjunto de datos.
# Se especifican diferentes argumentos para personalizar el gráfico.
# 'taxis': el conjunto de datos a utilizar.
# 'hue': variable categórica para diferenciar las observaciones con diferentes colores en el gráfico.
# 'vars': variables a incluir en el gráfico.
# 'palette': paleta de colores a utilizar para la variable categórica.
# 'markers': símbolos a utilizar para representar cada punto en el gráfico.
axis = sns.pairplot(taxis, 
                    hue="payment", 
                    vars=["fare", "tip"], 
                    palette="husl",
                    markers=["o", "s", "D"])
```

```python
sns.pairplot(data=taxis, hue="pickup_zone")
```
