```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import warnings
import seaborn as sns


```

```python
taxis = sns.load_dataset('titanic')

taxis
```

```python
x = taxis["distance"]
y = taxis["fare"]

sns.jointplot(x=x, y=y, data=taxis, kind="scatter", height=6, ratio=5, space=0.2, xlim=None, ylim=None, joint_kws=None, marginal_kws=None)

```

```python
# E mÃºltiples tipos de grÃ¡ficos para proporcionar rÃ¡pidamente resÃºmenes informativos de un conjunto de datos. Uno, jointplot(), se enfoca en una sola relaciÃ³n. 
#Traza la distribuciÃ³n conjunta entre dos variables junto con la distribuciÃ³n marginal de cada variable:

#sns.jointplot(data=taxis, x="distance", y="fare", hue="tip",xlim=[0:60], ylim=[0:200])
sns.jointplot(data=taxis, x="distance", y="fare", hue="tip")

```

```python
# Se importa la librerÃ­a Seaborn para hacer grÃ¡ficos estadÃ­sticos
import seaborn as sns

# Se utiliza la funciÃ³n pairplot de Seaborn para crear un grÃ¡fico de dispersiÃ³n de todas las variables numÃ©ricas en un conjunto de datos.
# Se especifican diferentes argumentos para personalizar el grÃ¡fico.
# 'taxis': el conjunto de datos a utilizar.
# 'hue': variable categÃ³rica para diferenciar las observaciones con diferentes colores en el grÃ¡fico.
# 'vars': variables a incluir en el grÃ¡fico.
# 'palette': paleta de colores a utilizar para la variable categÃ³rica.
# 'markers': sÃ­mbolos a utilizar para representar cada punto en el grÃ¡fico.
axis = sns.pairplot(taxis, 
                    hue="payment", 
                    vars=["fare", "tip"], 
                    palette="husl",
                    markers=["o", "s", "D"])
```
