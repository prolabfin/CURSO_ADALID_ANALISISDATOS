```python
# importamos la librería seaborn y la renombramos como sns
import seaborn as sns

# cargamos el dataset "tips" que viene por defecto en seaborn
tips = sns.load_dataset("tips")

# definimos la variable x como la columna "total_bill" del dataset tips
x = tips["total_bill"]

# definimos la variable y como la columna "tip" del dataset tips
y = tips["tip"]

# generamos un jointplot con los datos x e y, de tipo "scatter" que muestra la relación entre ambas variables
# height es la altura de la gráfica en pulgadas
# ratio es la proporción entre las dimensiones x e y del jointplot
# space es el espacio en blanco que se agrega entre las gráficas de dispersión y las marginales
# xlim y ylim son los límites del eje x e y respectivamente
# joint_kws son los parámetros de configuración de la gráfica de dispersión, en este caso no se ha especificado ninguno
# marginal_kws son los parámetros de configuración de las gráficas marginales, en este caso no se ha especificado ninguno


sns.jointplot(
    x=x,                     # variable para el eje x
    y=y,                     # variable para el eje y
    data=tips,               # dataset a utilizar
    kind="scatter",          # tipo de gráfico a realizar, en este caso un scatter plot
    height=6,                # alto de la figura en pulgadas
    ratio=5,                 # relación entre el tamaño del scatter plot y los histogramas
    space=0.2,               # separación entre el scatter plot y los histogramas
    xlim=None,               # límites del eje x, en este caso se dejan por defecto
    ylim=None,               # límites del eje y, en este caso se dejan por defecto
    joint_kws=None,          # parámetros adicionales para el scatter plot
    marginal_kws=None,       # parámetros adicionales para los histogramas
)

```

```python
# Se crea un jointplot que muestra la relación entre la longitud del pétalo y la longitud del sépalo
# x e y indican las variables a graficar
# data indica el dataset que se utilizará para graficar las variables
# marginal_kws son los argumentos de estilo para los histogramas en los bordes de la gráfica
# s es el tamaño de los puntos en la gráfica
# edgecolor es el color del borde de los puntos
# linewidth es el ancho del borde de los puntos
ax = sns.jointplot(
    x="petal_length",
    y="sepal_length",
    data=iris,
    marginal_kws=dict(bins=15, rug=True),
    s=50, edgecolor="w", linewidth=1)

```

```python
# Se importa la librería Seaborn para hacer gráficos estadísticos
import seaborn as sns

# Se utiliza la función pairplot de Seaborn para crear un gráfico de dispersión de todas las variables numéricas en un conjunto de datos.
# Se especifican diferentes argumentos para personalizar el gráfico.
# 'data': el conjunto de datos a utilizar.
# 'hue': variable categórica para diferenciar las observaciones con diferentes colores en el gráfico.
# 'hue_order': orden de la variable categórica.
# 'palette': paleta de colores a utilizar para la variable categórica.
# 'vars': variables a incluir en el gráfico.
# 'x_vars': variables a incluir en el eje x del gráfico.
# 'y_vars': variables a incluir en el eje y del gráfico.
# 'kind': tipo de gráfico a utilizar (por defecto 'scatter', que es un gráfico de dispersión).
# 'diag_kind': tipo de gráfico para la diagonal principal del gráfico (por defecto 'auto', que escoge automáticamente el tipo de gráfico según el tipo de variable).
# 'markers': símbolos a utilizar para representar cada punto en el gráfico.
# 'height': altura de cada subgráfico en pulgadas.
# 'aspect': relación de aspecto de cada subgráfico.
# 'corner': si es True, sólo se muestran los gráficos de la diagonal principal y la esquina inferior izquierda.
# 'dropna': si es True, se eliminan las observaciones con valores faltantes.
# 'plot_kws': argumentos adicionales a pasar a la función de trazado (para cada subgráfico).
# 'diag_kws': argumentos adicionales a pasar a la función de trazado para la diagonal principal del gráfico.
# 'grid_kws': argumentos adicionales a pasar a la función que dibuja la cuadrícula.
# 'size': antiguo argumento de seaborn que ya no se utiliza.
sns.pairplot(iris, 
             hue=None, 
             hue_order=None,
             palette=None, 
             vars=None, 
             x_vars=None, 
             y_vars=None, 
             kind='scatter', 
             diag_kind='auto', 
             markers=None, 
             height=2.5, 
             aspect=1, 
             corner=False, 
             dropna=True, 
             plot_kws=None, 
             diag_kws=None, 
             grid_kws=None, 
             size=None)
```

```python
# Se importa la librería Seaborn para hacer gráficos estadísticos
import seaborn as sns

# Se utiliza la función pairplot de Seaborn para crear un gráfico de dispersión de todas las variables numéricas en un conjunto de datos.
# Se especifican diferentes argumentos para personalizar el gráfico.
# 'iris': el conjunto de datos a utilizar.
# 'hue': variable categórica para diferenciar las observaciones con diferentes colores en el gráfico.
# 'vars': variables a incluir en el gráfico.
# 'palette': paleta de colores a utilizar para la variable categórica.
# 'markers': símbolos a utilizar para representar cada punto en el gráfico.
axis = sns.pairplot(iris, 
                    hue="species", 
                    vars=["sepal_width", "sepal_length"], 
                    palette="husl",
                    markers=["o", "s", "D"])
```

```python
# Se utiliza la función barplot de Seaborn para crear un gráfico de barras que muestra la media (o mediana) de una variable numérica para diferentes categorías.
# Se especifican diferentes argumentos para personalizar el gráfico.
# 'x': la variable categórica del eje x en el gráfico.
# 'y': la variable numérica del eje y en el gráfico.
# 'hue': variable categórica adicional para diferenciar las observaciones con diferentes colores en el gráfico.
# 'order': orden de las categorías en el eje x del gráfico.
# 'data': el conjunto de datos a utilizar.
# 'estimator': estadístico a utilizar para resumir los datos dentro de cada categoría ('np.median' en este caso).
# 'ci': nivel de intervalo de confianza para trazar en el gráfico.
# 'n_boot': número de veces que se muestran los intervalos de confianza.
# 'palette': paleta de colores a utilizar para la variable categórica.
# 'saturation': saturación del color de las barras.
# 'errcolor': color de las líneas de error.
# 'errwidth': grosor de las líneas de error.
# 'capsize': longitud de las líneas en las barras que representan el intervalo de confianza.
# 'dodge': separación de las barras para las diferentes categorías.
# 'ax': el objeto de ejes de matplotlib donde se trazará el gráfico.
ax = sns.barplot(x="day", 
                 y="total_bill", 
                 hue="sex", 
                 order=["Sun", "Sat", "Thur", "Fri"], 
                 data=tips, 
                 estimator=np.median, 
                 ci=95, 
                 n_boot=1000, 
                 units=None, 
                 orient=None, 
                 color=None, 
                 palette="Set3", 
                 saturation=0.75, 
                 errcolor="c", 
                 errwidth=1, 
                 capsize=0.2, 
                 dodge=True, 
                 ax=None)

# Se agrega una leyenda al gráfico
plt.legend(loc="best")
# Se agrega un título al gráfico
plt.title("seaborn.barplot")
# Se muestra el gráfico
plt.show()
```

```python
# Se importa la librería Seaborn para hacer gráficos estadísticos
import seaborn as sns

# Se utiliza la función boxplot de Seaborn para crear un gráfico de caja que muestra la distribución de los datos.
# Se especifican diferentes argumentos para personalizar el gráfico.
# 'x': variable categórica o cuantitativa del eje x en el gráfico.
# 'y': variable categórica o cuantitativa del eje y en el gráfico.
# 'hue_order': orden de los niveles de la variable categórica adicional especificada en 'hue'.
# 'order': orden de los niveles de la variable categórica especificada en 'x' o 'y'.
# 'data': el conjunto de datos a utilizar.
# 'ci': nivel de confianza para los intervalos de confianza. 
# 'n_boot': número de iteraciones para el cálculo del intervalo de confianza.
# 'units': variable categórica adicional para agrupar los datos en el cálculo de los estadísticos de la caja.
# 'orient': orientación del gráfico ('v' para vertical, 'h' para horizontal).
# 'color': color de las cajas en el gráfico.
# 'palette': paleta de colores a utilizar.
# 'saturation': saturación de los colores de la paleta.
# 'width': ancho relativo de las cajas.
# 'dodge': si es True, separa las cajas de diferentes niveles de la variable categórica en el gráfico.
# 'fliersize': tamaño de los puntos que se muestran como valores atípicos.
# 'whis': factor de ampliación de los bigotes.
# 'ax': objeto de ejes Matplotlib en el que se grafica.
# '**kwargs': argumentos adicionales que se pasan a la función Matplotlib.boxplot().

# Se define el diccionario kwargs con argumentos de palabras clave adicionales
#kwargs = {'meanline': True, 'showmeans': True}

# Se crea un boxplot utilizando el dataset 'tips'
sns.boxplot(x='day',
            y='total_bill',
            hue='smoker',
            order=["Thur", "Fri", "Sat", "Sun"],
            hue_order=["No", "Yes"],
            data=tips,
            palette="Set2",
            saturation=0.75,
            width=0.8,
            dodge=True,
            fliersize=5,
            whis=1.5,
            ax=None)

# Se muestra el gráfico
plt.legend(loc="best")
plt.title("seaborn.boxplot")
plt.show()


```

```python
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_theme()

# Generate an example radial datast
r = np.linspace(0, 10, num=100)
df = pd.DataFrame({'r': r, 'slow': r, 'medium': 2 * r, 'fast': 4 * r})

# Convert the dataframe to long-form or "tidy" format
df = pd.melt(df, id_vars=['r'], var_name='speed', value_name='theta')

# Set up a grid of axes with a polar projection
g = sns.FacetGrid(df, col="speed", hue="speed",
                  subplot_kws=dict(projection='polar'), height=4.5,
                  sharex=False, sharey=False, despine=False)

# Draw a scatterplot onto each axes in the grid
g.map(sns.scatterplot, "theta", "r")
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

# Load the dataset
crashes = sns.load_dataset("car_crashes")

# Make the PairGrid
g = sns.PairGrid(crashes.sort_values("total", ascending=False),
                 x_vars=crashes.columns[:-3], y_vars=["abbrev"],
                 height=10, aspect=.25)

# Draw a dot plot using the stripplot function
g.map(sns.stripplot, size=10, orient="h", jitter=False,
      palette="flare_r", linewidth=1, edgecolor="w")

# Use the same x axis limits on all columns and add better labels
g.set(xlim=(0, 25), xlabel="Crashes", ylabel="")

# Use semantically meaningful titles for the columns
titles = ["Total crashes", "Speeding crashes", "Alcohol crashes",
          "Not distracted crashes", "No previous crashes"]

for ax, title in zip(g.axes.flat, titles):

    # Set a different title for each axes
    ax.set(title=title)

    # Make the grid horizontal instead of vertical
    ax.xaxis.grid(False)
    ax.yaxis.grid(True)

sns.despine(left=True, bottom=True)
```
