```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el conjunto de datos Titanic
titanic = sns.load_dataset("titanic")

# Filtrar los datos para obtener solo los supervivientes
survived = titanic[titanic['survived'] == 1]

# Crear el gráfico de barras
sns.countplot(x='sex', data=survived, palette='Blues')

# Personalizar el gráfico
plt.title('Sobrevivientes del Titanic por género')
plt.xlabel('Género')
plt.ylabel('Cantidad de sobrevivientes')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el conjunto de datos Titanic
titanic = sns.load_dataset("titanic")

# Filtrar los datos para obtener solo los supervivientes
survived = titanic[titanic['survived'] == 1]
deceased = titanic[titanic['survived'] == 0]

# Crear el gráfico de barras agrupados
sns.countplot(x='sex', hue='survived', data=titanic, palette='Blues')

# Personalizar el gráfico
plt.title('Sobrevivientes y fallecidos del Titanic por género')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.legend(title='Sobrevivió', loc='upper right', labels=['No', 'Sí'])
plt.show()
```
